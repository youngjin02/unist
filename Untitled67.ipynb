```python
import torch
import pandas as pd
import numpy as np
import json
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# NV-Embed-v2 모델 로드 (fp16 활성화)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = SentenceTransformer('nvidia/NV-Embed-v2', trust_remote_code=True)
model = model.half()  # fp16 설정
model = model.to(device)  # 모델을 GPU로 이동
model.max_seq_length = 32768  # NV-Embed-v2 최대 토큰 길이
model.tokenizer.padding_side = "right"  # 토큰화 방향 설정

# 데이터 전처리 함수
def validate_and_truncate_texts(texts, max_tokens):
    """텍스트 데이터에서 유효하지 않은 텍스트를 제거하고 최대 길이에 맞게 자름."""
    valid_texts = []
    for text in texts:
        if text and text.strip():  # 빈 텍스트 제거
            encoded = model.tokenizer.encode(text, truncation=True, max_length=max_tokens)
            truncated_text = model.tokenizer.decode(encoded)
            if truncated_text.strip():  # 자른 뒤에도 유효한 텍스트만 추가
                valid_texts.append(truncated_text)
            else:
                print(f"Warning: Truncated text is empty. Original text: {text}")
    return valid_texts

# 데이터 파일 경로
query_files = [
    '/content/queries.jsonl',
    '/content/queries2.jsonl',
    '/content/queries3.jsonl',
    '/content/queries4.jsonl',
    '/content/queries5.jsonl',
    '/content/queries6.jsonl',
    '/content/queries7.jsonl'
]

corpus_files = [
    '/content/corpus.jsonl',
    '/content/corpus2.jsonl',
    '/content/corpus3.jsonl',
    '/content/corpus4.jsonl',
    '/content/corpus5.jsonl',
    '/content/corpus6.jsonl',
    '/content/corpus7.jsonl'
]

# 결과를 저장할 리스트
final_results = []

# 각 쿼리와 코퍼스 파일을 처리
for query_file, corpus_file in zip(query_files, corpus_files):
    print(f"Processing {query_file} and {corpus_file}...")

    # 쿼리와 코퍼스 파일 읽기
    with open(query_file, 'r', encoding='utf-8') as f:
        queries = [json.loads(line) for line in f]

    with open(corpus_file, 'r', encoding='utf-8') as f:
        corpus = [json.loads(line) for line in f]

    # 코퍼스 텍스트와 ID 추출 및 전처리
    corpus_texts = [doc['text'] for doc in corpus]
    corpus_texts = validate_and_truncate_texts(corpus_texts, max_tokens=model.max_seq_length)
    corpus_ids = [doc['_id'] for doc in corpus]

    # 코퍼스 임베딩 계산
    print("코퍼스 임베딩 계산 중...")
    corpus_embeddings = model.encode(
        corpus_texts,
        batch_size=1,
        normalize_embeddings=True,
        device=device
    )

    # 각 쿼리에 대해 유사도 계산 및 상위 10개의 코퍼스 선택
    for query in queries:
        query_id = query['_id']
        query_text = validate_and_truncate_texts([query['text']], max_tokens=model.max_seq_length)[0]

        # 쿼리 텍스트 임베딩 생성
        print(f"쿼리 임베딩 생성 중... (query_id: {query_id})")
        query_embedding = model.encode(
            [query_text],
            normalize_embeddings=True,
            device=device
        )[0]

        # 코사인 유사도 계산 전 NaN 검사 및 제거
        if np.isnan(query_embedding).any():
            print(f"Query embedding contains NaN values for query_id: {query_id}")
            query_embedding = np.nan_to_num(query_embedding)

        if np.isnan(corpus_embeddings).any():
            print(f"Corpus embeddings contain NaN values. Removing NaNs...")
            corpus_embeddings = np.nan_to_num(corpus_embeddings)

        # 코사인 유사도 계산
        similarities = cosine_similarity([query_embedding], corpus_embeddings).flatten()

        # 상위 10개의 코퍼스 인덱스를 선택
        top_indices = similarities.argsort()[-10:][::-1]

        # 상위 10개의 결과 저장
        for idx in top_indices:
            final_results.append({
                'query_id': query_id,
                'corpus_id': corpus_ids[idx]
            })

# 최종 결과를 DataFrame으로 변환
results_df = pd.DataFrame(final_results)

# 결과를 CSV 파일로 저장
final_output_file = '/content/final_submission_nv_embed_v2_fp16_combined.csv'
results_df.to_csv(final_output_file, index=False)
print(f"결과가 {final_output_file}에 저장되었습니다.")
```
